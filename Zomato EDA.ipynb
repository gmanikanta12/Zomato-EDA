```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("zomato.csv",encoding='latin-1')
df.head()
```

```python
# to  get null values count 
df.info()
```

## In Data Analysis what all thimngs we do is
1. Missing values
2. explore about numerical values
3. explore about categorical values
4. find relationships between features

```python
# to find the null values in features 
df.isnull().sum()
```

```python
#other way to find null values
[features for features in df.columns if df[features].isnull().sum()>0]
```

```python
# to find null values by heatmap
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
# now we will go to the country code data
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
# now we merge the data sets by using merge function in pandas.
# by using this we will add country column to df 
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head()
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_names
```

```python
country_val=final_df.Country.value_counts().values
```

```python
country_val
```

```python
## now we will go for pie chart with country_names and values
plt.pie(country_val[:3],labels=country_names[:3],autopct="%1.2f%%")
```

Observation:
Zomato mostly operates in india. After that  US, Uk

```python
# now one more observation with aggregated ratings
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
# now we rename the last column as rating count 
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:"rating count"})
```

```python
ratings
```

## Observation
1. when ration is between 4.5 to 4.9 ---> Excellent
2. when ration is between 4.0 to 4.4 ---> Very good
3. when ration is between 3.5 to 3.9 ---> good


```python
# if we plot a bar graph then we will see this.
plt.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating", y="rating count",data=ratings)
```

```python
sns.barplot(x="Aggregate rating", y="rating count",hue='Rating color',data=ratings,palette=['black','red','Orange','yellow','green','blue'])
```

Observation:
  1. Not rated count is tto high.
  2. Maximum number of rating are between 2.5 to 3.4

```python
# now we draw a count plot for raings
sns.countplot(x="Rating color",data=ratings,palette=['black','red','Orange','yellow','green','blue'])
```

```python
# find countries with 0 ratings given
final_df[final_df["Aggregate rating"]==0.0].groupby('Country').size().reset_index()
```
